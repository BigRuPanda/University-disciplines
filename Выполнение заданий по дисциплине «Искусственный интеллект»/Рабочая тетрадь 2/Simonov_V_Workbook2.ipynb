```python
# Задание 2.3.1

import pandas as pd
from math import sqrt

a = pd.Series([1, 2, 3, 4, 5])
b = pd.Series([10, 20, 30, 40, 50])


def eRas(y, z):
  res = 0

  for i in range(len(y)):
    res += (y[i] - z[i]) ** 2

  return sqrt(res)

print(eRas(a, b))
```
