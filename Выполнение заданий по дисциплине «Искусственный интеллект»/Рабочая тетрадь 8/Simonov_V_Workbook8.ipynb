```python
# Задание 1

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.datasets import load_digits
sns.set()

X = np.array([
    [5,3],
    [10,15],
    [15,12],
    [24,10],
    [30,45],
    [85,70],
    [71,80],
    [60,78],
    [55,52],
    [80,91],
])

fig, ax = plt.subplots(1, 4, figsize=(12, 4))

for axi, i in zip(ax.flat, range(1, 5)):
    kmeans = KMeans(n_clusters=i, random_state=0, n_init=10)
    kmeans.fit(X)
    y = kmeans.predict(X)
    axi.scatter(X[:, 0], X[:, 1], c=y, cmap = "autumn")
    axi.set_title("Кол-во кластеров: " + str(i))
```

```python
# Задание 2

iris = pd.read_csv('https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv')
x_iris = iris.iloc[:, :-1]
y_iris = iris.iloc[:, -1]


kmeans = KMeans(n_clusters=8, n_init=10, random_state=3)
n = kmeans.fit_predict(x_iris)

sns.scatterplot(
    data = iris,
    x='petal.length',
    y='petal.width',
    c = n,
    cmap = "cool"
    ).set_title('Кол-во кластеров: 8')

fig, ax = plt.subplots(1, 3, figsize=(12, 4))

for axi, i in zip(ax.flat, range(2, 5)):

    kmeans = KMeans(n_clusters=i, n_init=10, random_state=3)
    n = kmeans.fit_predict(x_iris)

    axi.scatter(
    data = iris,
    x='petal.length',
    y='petal.width',
    c = n,
    cmap = "autumn"
    )

    axi.set_title("Кол-во кластеров: " + str(i))
```
