```python
# Задание 1.3.1

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

def sqr_euclidean_distanse(v1, v2):
    return sum((x - y) ** 2 for x, y in zip(v1, v2))
def weighted_euclidean_distanse(v1, v2, w):
    return sum((x - y) ** 2 * s for (x, y, s) in zip(v1, v2, w))**0.5
def manhattan_distanse(v1, v2):
    return sum(abs(x - y) for (x, y) in zip(v1, v2))
def cheb_distanse(v1, v2):
    return max(abs(x - y) for (x, y) in zip(v1, v2))

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

x = np.array([0,6,4])
y = np.array([3,7,1])
w = np.array([4,0,4])
v = np.array([0,0,2])

ax.scatter(0,6,4)
ax.scatter(3,7,1)
ax.scatter(4,0,4)
ax.scatter(0,0,2)

print("Расстояние Евклида = ",np.linalg.norm(x-y))
print("Расстояние Евклида = ",np.linalg.norm(x-v))
print("Расстояние Евклида = ",np.linalg.norm(x-w))
print("Расстояние Евклида = ",np.linalg.norm(v-w))
print("Расстояние Евклида = ",np.linalg.norm(y-v))

print("Квадрат Евклидова расстояния = ",np.linalg.norm(x-y) ** 2)
print("Квадрат Евклидова расстояния = ",np.linalg.norm(x-v) ** 2)
print("Квадрат Евклидова расстояния = ",np.linalg.norm(x-w) ** 2)
print("Квадрат Евклидова расстояния = ",np.linalg.norm(v-w) ** 2)
print("Квадрат Евклидова расстояния = ",np.linalg.norm(y-v) ** 2)

print("Расстояние Чебышева = ",np.linalg.norm(x-y, ord=np.inf))
print("Расстояние Чебышева = ",np.linalg.norm(x-v, ord=np.inf))
print("Расстояние Чебышева = ",np.linalg.norm(x-w, ord=np.inf))
print("Расстояние Чебышева = ",np.linalg.norm(v-w, ord=np.inf))
print("Расстояние Чебышева = ",np.linalg.norm(y-v, ord=np.inf))

print("Расстояние Хемминга = ",np.linalg.norm(x-y, ord = 1))
print("Расстояние Хемминга = ",np.linalg.norm(x-v, ord = 1))
print("Расстояние Хемминга = ",np.linalg.norm(x-w, ord = 1))
print("Расстояние Хемминга = ",np.linalg.norm(v-w, ord = 1))
print("Расстояние Хемминга = ",np.linalg.norm(y-v, ord = 1))

plt.show()
```

```python
# Задание 2.3.1

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D

iris = sns.load_dataset('iris')

X_train, X_test, y_train, y_test = train_test_split(
    iris.iloc[:,:-1],
    iris.iloc[:,-1],
    test_size = 0.15
)
X_train.head()
y_train.head()
model = KNeighborsClassifier(n_neighbors = 10)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
plt.figure(figsize = (10,7))
sns.scatterplot(
    data = iris,
    x = 'petal_width', y = 'petal_length',
    hue = 'species',
    s = 70
)
plt.xlabel('Длина лепестка, см')
plt.ylabel('Ширина лепестка, см')
plt.legend(loc = 2)
plt.grid()

for i in range(len(y_test)):
    if np.array(y_test)[i] != y_pred[i]:
        plt.scatter(X_test.iloc[i, 3],X_test.iloc[i, 2], color = 'red', s = 150)
from sklearn.metrics import accuracy_score
print(f'accuracy: {accuracy_score(y_test,y_pred):.3}')
```
