```python
#Задание 4.3.1

from random import uniform
from statistics import median
import matplotlib.pyplot as plt

a = [uniform(0, 1) for i in range(10)]
sr = sum(a)/len(a)
med = median(a)

print('Среднее арифметическое:', sr, '\nМедиана:', med)
plt.grid()
plt.scatter(a, a, label = 'Элементы массива')
plt.scatter(sr, sr, label = 'Среднее арифметическое')
plt.scatter(med, med, label = 'Медиана')
plt.legend()
```

```python
#Задание 4.3.2

import math as m
def f1(x):
  return (m.sqrt(1 + m.exp(m.sqrt(x)) + m.cos(x ** 2)) / abs(1 - m.sin(x) ** 3)) + m.log(abs(2 * x))

a = [f1(x) for x in range(1,11)]
print(a)

plt.grid()
plt.plot([i for i in range(len(a))], a)
plt.show()
plt.grid()
plt.scatter(a[0:5], a[0:5])
plt.show()
```

```python
#Задание 4.3.3

import numpy as np
from scipy.integrate import simps
from numpy import trapz

def f2(x):
  return abs(np.cos(x * np.exp(m.cos(x) + np.log(x+1))))

y = [f2(i) for i in range(0,10)]

plt.grid()
plt.plot([i for i in range(0,10)],y, c = 'r')
plt.fill_between( [i for i in range(0,10)],y)

area = trapz(y)
print(area)
```

```python
#Задание 4.3.5

from math import *

inp = str(input('Введите функцию "+", "-", "*", "/", "^", "e" = e^(x=y), "sin" = sin(x+y), "cos" = cos(x+y) => '))
x = int(input('Введите x: '))
y = int(input('Введите y: '))

if inp == '+':
  print(x+y)
elif inp == '-':
  print(x-y)
elif inp == '*':
  print(x*y)
elif inp == '/':
  if (y != 0):
    print(x/y)
  else:
    print('!!! Деление на 0 !!!')
elif inp == '^':
  print(x**y)
elif inp == 'e':
  print(e**(x+y))
elif inp == 'sin':
  print(sin(x+y))
elif inp == 'cos':
  print(cos(x+y))
```
